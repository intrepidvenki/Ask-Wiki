```python
import os
import requests
import shutil
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re  # Import re to clean filenames

# Function to sanitize filenames
def clean_filename(filename):
    return re.sub(r'[\\/*?:"<>|]', "_", filename)

# Function to download images
def download_image(image_url, folder):
    if not os.path.exists(folder):
        os.makedirs(folder)
    
    # Clean the filename to avoid invalid characters
    image_name = clean_filename(image_url.split("/")[-1])
    file_name = os.path.join(folder, image_name)

    response = requests.get(image_url, stream=True)
    if response.status_code == 200:
        with open(file_name, 'wb') as out_file:
            shutil.copyfileobj(response.raw, out_file)
        del response
        print(f"Image downloaded: {file_name}")
    else:
        print(f"Failed to download image: {image_url}")

# Function to extract text and images
def extract_text_and_images(url, folder='images'):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract text
    text = soup.get_text(separator="\n")
    print("Text content:")
    print(text)

    # Extract and download images
    images = soup.find_all('img')
    for img in images:
        img_url = img.get('src')
        img_url = urljoin(url, img_url)  # Handle relative URLs
        download_image(img_url, folder)

url = 'https://en.wikipedia.org/wiki/World_War_II'
extract_text_and_images(url)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests
import os
import shutil
from urllib.parse import urljoin

# Function to download images
def download_image(image_url, folder):
    if not os.path.exists(folder):
        os.makedirs(folder)
    response = requests.get(image_url, stream=True)
    if response.status_code == 200:
        file_name = os.path.join(folder, image_url.split("/")[-1])
        with open(file_name, 'wb') as out_file:
            shutil.copyfileobj(response.raw, out_file)
        del response
        print(f"Image downloaded: {file_name}")

# Function to extract text and images using Selenium
def extract_text_and_images(url, folder='images'):
    # Set up Selenium WebDriver
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(url)

    # Get page source and parse it with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Extract text
    text = soup.get_text(separator="\n")
    print("Text content:")
    print(text)

    # Extract and download images
    images = soup.find_all('img')
    for img in images:
        img_url = img.get('src')
        img_url = urljoin(url, img_url)  # Handle relative URLs
        download_image(img_url, folder)

    driver.quit()

url = 'https://en.wikipedia.org/wiki/World_War_II'
extract_text_and_images(url)
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin, urlparse

def extract_webpage_content(url):
    # Send a GET request to the webpage
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract all text from the webpage
    text = soup.get_text(separator='\n', strip=True)
    
    # Extract all images from the webpage
    images = []
    for img in soup.find_all('img'):
        img_url = img.get('src')
        if img_url:
            # Convert relative URLs to absolute URLs
            img_url = urljoin(url, img_url)
            images.append(img_url)
    
    return text, images

def save_content(url, text, images):
    # Create a directory to store the extracted content
    domain = urlparse(url).netloc
    directory = f"{domain}_extracted"
    os.makedirs(directory, exist_ok=True)
    
    # Save the extracted text
    with open(os.path.join(directory, "extracted_text.txt"), "w", encoding="utf-8") as f:
        f.write(text)
    
    # Save the images
    for i, img_url in enumerate(images):
        try:
            img_data = requests.get(img_url).content
            filename = f"image_{i+1}.{img_url.split('.')[-1]}"
            with open(os.path.join(directory, filename), "wb") as f:
                f.write(img_data)
        except Exception as e:
            print(f"Error downloading image {img_url}: {e}")

def main():
    url = input("Enter the URL of the webpage you want to extract: ")
    text, images = extract_webpage_content(url)
    save_content(url, text, images)
    print(f"Extraction complete. Content saved in the '{urlparse(url).netloc}_extracted' directory.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin, urlparse
import re

def extract_webpage_content(url):
    # Send a GET request to the webpage
    response = requests.get(url)
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Remove script and style elements
    for script in soup(["script", "style"]):
        script.decompose()
    
    # Get the base URL for resolving relative URLs
    base_url = '{uri.scheme}://{uri.netloc}'.format(uri=urlparse(url))
    
    # Process all elements to convert relative URLs to absolute
    for tag in soup.find_all(['a', 'img', 'link', 'script']):
        if tag.has_attr('href'):
            tag['href'] = urljoin(base_url, tag['href'])
        if tag.has_attr('src'):
            tag['src'] = urljoin(base_url, tag['src'])
    
    # Extract all images
    images = [img['src'] for img in soup.find_all('img') if img.has_attr('src')]
    
    return soup, images

def save_content(url, soup, images):
    # Create a directory to store the extracted content
    domain = urlparse(url).netloc
    directory = f"{domain}_extracted"
    os.makedirs(directory, exist_ok=True)
    
    # Save the HTML content
    with open(os.path.join(directory, "index.html"), "w", encoding="utf-8") as f:
        f.write(str(soup))
    
    # Save the images
    for i, img_url in enumerate(images):
        try:
            img_data = requests.get(img_url).content
            filename = f"image_{i+1}.{img_url.split('.')[-1].split('?')[0]}"
            with open(os.path.join(directory, filename), "wb") as f:
                f.write(img_data)
            
            # Update image src in the HTML to point to local file
            for img_tag in soup.find_all('img', src=img_url):
                img_tag['src'] = filename
        except Exception as e:
            print(f"Error downloading image {img_url}: {e}")
    
    # Save the updated HTML content with local image references
    with open(os.path.join(directory, "index_local.html"), "w", encoding="utf-8") as f:
        f.write(str(soup))

def main():
    url = input("Enter the URL of the webpage you want to extract: ")
    soup, images = extract_webpage_content(url)
    save_content(url, soup, images)
    print(f"Extraction complete. Content saved in the '{urlparse(url).netloc}_extracted' directory.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin, urlparse
import re

def extract_webpage_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Remove script and style elements
    for script in soup(["script", "style"]):
        script.decompose()
    
    base_url = '{uri.scheme}://{uri.netloc}'.format(uri=urlparse(url))
    
    # Process all elements to convert relative URLs to absolute
    for tag in soup.find_all(['a', 'img', 'link', 'script']):
        if tag.has_attr('href'):
            tag['href'] = urljoin(base_url, tag['href'])
        if tag.has_attr('src'):
            tag['src'] = urljoin(base_url, tag['src'])
    
    images = [img['src'] for img in soup.find_all('img') if img.has_attr('src')]
    
    return soup, images

def extract_organized_text(soup):
    organized_text = []
    
    # Extract title
    title = soup.title.string if soup.title else "No Title"
    organized_text.append(f"Title: {title}\n")
    
    # Extract headers
    for i in range(1, 7):
        headers = soup.find_all(f'h{i}')
        if headers:
            organized_text.append(f"\nH{i} Headers:")
            for header in headers:
                organized_text.append(header.get_text(strip=True))
    
    # Extract paragraphs
    paragraphs = soup.find_all('p')
    if paragraphs:
        organized_text.append("\nParagraphs:")
        for p in paragraphs:
            organized_text.append(p.get_text(strip=True))
    
    # Extract lists
    lists = soup.find_all(['ul', 'ol'])
    if lists:
        organized_text.append("\nLists:")
        for lst in lists:
            items = lst.find_all('li')
            for item in items:
                organized_text.append(f"- {item.get_text(strip=True)}")
    
    # Extract links
    links = soup.find_all('a')
    if links:
        organized_text.append("\nLinks:")
        for link in links:
            organized_text.append(f"{link.get_text(strip=True)} : {link.get('href')}")
    
    return "\n".join(organized_text)

def save_content(url, soup, images, organized_text):
    domain = urlparse(url).netloc
    directory = f"{domain}_extracted"
    os.makedirs(directory, exist_ok=True)
    
    # Save the original HTML content
    with open(os.path.join(directory, "index.html"), "w", encoding="utf-8") as f:
        f.write(str(soup))
    
    # Save the organized text
    with open(os.path.join(directory, "organized_text.txt"), "w", encoding="utf-8") as f:
        f.write(organized_text)
    
    # Save the images
    for i, img_url in enumerate(images):
        try:
            img_data = requests.get(img_url).content
            filename = f"image_{i+1}.{img_url.split('.')[-1].split('?')[0]}"
            with open(os.path.join(directory, filename), "wb") as f:
                f.write(img_data)
            
            # Update image src in the HTML to point to local file
            for img_tag in soup.find_all('img', src=img_url):
                img_tag['src'] = filename
        except Exception as e:
            print(f"Error downloading image {img_url}: {e}")
    
    # Save the updated HTML content with local image references
    with open(os.path.join(directory, "index_local.html"), "w", encoding="utf-8") as f:
        f.write(str(soup))

def main():
    url = input("Enter the URL of the webpage you want to extract: ")
    soup, images = extract_webpage_content(url)
    organized_text = extract_organized_text(soup)
    save_content(url, soup, images, organized_text)
    print(f"Extraction complete. Content saved in the '{urlparse(url).netloc}_extracted' directory.")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urlparse

def extract_raw_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Remove script and style elements
    for element in soup(["script", "style"]):
        element.decompose()

    def extract_element(element, depth=0):
        if element.name is None:
            return element.strip()
        
        content = []
        if element.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            content.append(f"{'#' * depth} {element.get_text(strip=True)}")
        elif element.name == 'a':
            content.append(f"{element.get_text(strip=True)} [{element.get('href', '')}]")
        elif element.name == 'img':
            content.append(f"[Image: {element.get('alt', 'No description')}] ({element.get('src', '')})")
        elif element.name in ['ul', 'ol']:
            for li in element.find_all('li', recursive=False):
                content.append(f"{'  ' * depth}- {extract_element(li, depth + 1)}")
        elif element.name == 'table':
            content.append("Table:")
            for row in element.find_all('tr', recursive=False):
                content.append('  ' * (depth + 1) + ' | '.join(cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])))
        else:
            for child in element.children:
                child_content = extract_element(child, depth + 1)
                if child_content:
                    content.append(child_content)
        
        return '\n'.join(content)

    raw_content = extract_element(soup.body)
    
    # Extract title separately
    title = soup.title.string if soup.title else "No Title"
    
    return title, raw_content

def save_content(url, title, raw_content):
    domain = urlparse(url).netloc
    directory = f"{domain}_extracted"
    os.makedirs(directory, exist_ok=True)
    
    with open(os.path.join(directory, "raw_content.txt"), "w", encoding="utf-8") as f:
        f.write(f"Title: {title}\n\n")
        f.write(raw_content)

def main():
    url = input("Enter the URL of the webpage you want to extract: ")
    title, raw_content = extract_raw_content(url)
    save_content(url, title, raw_content)
    print(f"Extraction complete. Content saved in the '{urlparse(url).netloc}_extracted' directory.")

if __name__ == "__main__":
    main()
```
